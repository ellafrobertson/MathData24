```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
from IPython.display import Audio

# Generate x values and sine wave (sound wave)
x = np.linspace(-10, 10, 100)
y = np.sin(2 * x)

# Plot the sine wave
plt.plot(x, y)
plt.title("Generated Sound Wave")
plt.xlabel("x values")
plt.ylabel("Amplitude")
plt.show()

# Generating a sound wave with higher sample rate for audio (like a sine wave)
sampling_rate = 44100  # 44.1 kHz, common sampling rate for audio
t = np.linspace(0, 2, 2 * sampling_rate)  # Generate time values for 2 seconds
frequency = 440  # Frequency of the sine wave in Hz (A4 note)
audio_wave = 0.5 * np.sin(2 * np.pi * frequency * t)

# Write the wave to a WAV file (optional)
write("sound_wave.wav", sampling_rate, audio_wave)

# Play the audio in Colab
Audio(audio_wave, rate=sampling_rate)
```

```python
M = np.linspace(-10,10,20)
B = np.linspace(-10,10,20)
```

```python
for b in B:
  for m in M:
    print(m,b)
```

```python
best_loss = 1000000

for b in B:
  for m in M:

    m_guess = m
    b_guess = b
    y_guess = m_guess * x + b_guess

    loss = np.sum((y-y_guess)**2)

    if loss < best_loss:
      best_loss = loss
      print(m,b,loss)

      plt.plot(x,y,'b.')

      plt.plot(x,y_guess, 'r.')

      plt.show()
```
